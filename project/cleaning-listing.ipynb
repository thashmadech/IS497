```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

listing = pd.concat(li, axis=0, ignore_index=True)
pd.set_option('display.max_columns', None)
```

```python
listing.rename(index=str,columns={"name":"listing_name"},inplace=True)
listing.head()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(listing.dtypes)
```

```python
listing_df = listing[["id","listing_name","neighbourhood_cleansed",
                      "neighbourhood_group_cleansed","latitude","longitude",
                      "property_type","room_type",
                      "accommodates","bathrooms","bedrooms","beds","price",
                      "minimum_nights","maximum_nights","minimum_nights_avg_ntm","maximum_nights_avg_ntm",
                       "has_availability",
                      "availability_30","availability_60","availability_90","availability_365",
                      "number_of_reviews","first_review", "last_review",
                      "review_scores_rating","review_scores_accuracy","review_scores_cleanliness",
                      "review_scores_checkin","review_scores_communication","review_scores_location",
                      "review_scores_value","instant_bookable","reviews_per_month"]].copy()
```

```python
listing_df = listing_df.fillna(0)
```

```python
# convert dates to datetime
listing_df["first_review"]=pd.to_datetime(listing_df["first_review"])
listing_df["last_review"]=pd.to_datetime(listing_df["last_review"])
```

```python
# convert boolean columns to boolean

listing_df["instant_bookable"].replace(["t","f"],[True,False],inplace=True)
listing_df["has_availability"].replace(["t","f"],[True,False],inplace=True)
```

```python
listing_df['price'] = listing_df['price'].replace({'\$': '', ',': ''}, regex=True)
```

```python
listing_df["price"] = pd.to_numeric(listing_df["price"])
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(listing_df.dtypes)
```

```python
listing_df.to_csv("listing.csv",index=False) 
```
